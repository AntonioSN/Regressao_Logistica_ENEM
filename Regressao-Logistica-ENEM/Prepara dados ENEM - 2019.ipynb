# Prepara dados ENEM 2019

 Esse código lê os microdados ENEM e faz um tratamento preliminar como:
- Tratamento de valores perdidos
- Remoção de informações redundantes

Optou-se por tomar os códigos.

# Proporção de alunos que optaram por recurso especial para cada tipo de recurso

```python
M = len(DF)
DF[REE].sum().apply(lambda x: x/M)
```

# Proporção de alunos que optaram por pelo menos um recurso especial

```python
#Contaa o número de alunos que optaram por pelomentos um recurso especial.
Total_REE = DF[REE].apply(lambda x: x.sum(), axis=1).apply(lambda x: 0 if (x!=0) else 1).sum()

#Calcula o percentual de alunos que optaram por pelo menos um recurso especial.
print('Total de alunos na amostra: ',M)
print('Total de alunos que solicitaram pelo menos um recurso especial é de: ',Total_REE)
print('Proporção de alunos que optaram por pelo menos um recurso especial: ',Total_REE/M)
```

```python
DF['RECURSO_ESPECIAL'] = DF[REE].apply(lambda x: x.sum(), axis=1).apply(lambda x: 1 if (x!=0) else 0)

DF = DF.drop(columns=REE)
```

# Idade

Para que a regularização dos valores de idade não fique muito discrepantes com idades altas e pouco frequentes na amostra idades superiores a 25 anos foram concideradas outliers.

A seguir um histogramaa dos valores e a contagem dos valores.

# Q001
## Até que série seu pai, ou o homem responsável por você, estudou?

A	Nunca estudou.

B	Não completou a 4ª série/5º ano do Ensino Fundamental.

C	Completou a 4ª série/5º ano, mas não completou a 8ª série/9º ano do Ensino Fundamental.

D	Completou a 8ª série/9º ano do Ensino Fundamental, mas não completou o Ensino Médio.

E	Completou o Ensino Médio, mas não completou a Faculdade.

F	Completou a Faculdade, mas não completou a Pós-graduação.

G	Completou a Pós-graduação.

H	Não sei.


Como a maioria das categorias são ordinais uma única variável onde 0 representa "Nunca estudou" e 1 representa "completou a Pós-graduação". Respostas marcadas como "Não sei." serão consideradas de valor zero.

A categoria "Estudo_Pai" pode ser entendida como a ciência do estudante da escolaridade do pai até determinado grau onde:

### 0: Não sabe ou o pai nunca estudou.
### 1: Pai completou a pós graduação.

```python
DF['Estudo_Pai'] = DF['Q001'].replace({'A':0,
                                       'B':1/6,
                                       'C':2/6,
                                       'D':3/6,
                                       'E':4/6,
                                       'F':5/6,
                                       'G':6/6,
                                       'H':0,
                                      })
DF = DF.drop(columns = 'Q001')
```
