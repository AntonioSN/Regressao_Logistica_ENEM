```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Abrir o arquivo TXT
nome_arquivo = "MICRODADOS_ENEM_2021.csv" 
with open(nome_arquivo, "r") as arquivo:
    # Ler as linhas do arquivo
    linhas = arquivo.readlines()


m=len(linhas)
permuteIndex = np.random.permutation(np.arange(m))
Npart = 300
r = m % Npart
intervalos = int((m-r)/Npart)

df = pd.DataFrame([x.strip().split(';') for x in linhas[1:(intervalos+r)]], columns = linhas[0].strip().split(';'))

df.to_csv(nome_arquivo.split('.')[0]+"_1-"+str(Npart)+'.'+nome_arquivo.split('.')[1],index=False)
```
