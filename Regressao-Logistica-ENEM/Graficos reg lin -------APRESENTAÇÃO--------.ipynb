```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LogisticRegression, LinearRegression

def reforca(v,n):
    c = v
    for i in range(n):
        c = np.append(c,v)
    return c


%matplotlib notebook
```

```python
# Dados de exemplo
#Nota do aluno
y = np.array([ 3,2.8,4.8,5.5,6.1,5,6.8,7.5,8.2,10])

# Horas de estudo diário
x_1 = np.array([0.2,0.3,0.5,1.5,1.5,2.1,2.4,3.3,3.5,3.9])

# Quantidades de advertência ao longo do curso
x_2 = np.array([9,7,8,5,5,3,2,1,0,0])

#aprovado
x_3 = [0,0,0,0,1,0,1,1,1,1]
```

```python

header = r"""\begin{table}[h!]
\begin{adjustbox}{width=0.9\textwidth,center}
\begin{tabular}{|c|c|c|c|}
\hline
\thead{\textbf{Aluno}} &\thead{\textbf{$y$} \\ \textbf{nota na avaliação}} & \thead{\textbf{$x_1$} \\ \textbf{horas de estudo}} & \thead{\textbf{$x_2$} \\ \textbf{número de advertências}} \\ \hline
"""

footer = r"""
\end{tabular}
\end{adjustbox}
\end{table}
"""

table_body = ""
for i, yi, x1i, x2i in zip(range(len(y)),y, x_1, x_2):
    table_body += f"{i+1} & {yi} & {x1i} & {x2i} \\\ \hline\n"

latex_table = header + table_body + footer

print(latex_table)
```

```python

# Regressão linear para os dados 2D:
lr2D = LinearRegression()
lr2D.fit(x_1.reshape(-1, 1), y)
y_pred_2D = lr2D.predict(x_1.reshape(-1, 1))

# Plotando a regressão linear 2D:
plt.figure(figsize=(9,4))
plt.scatter(x_1, y, color='blue', label='Dados')
plt.plot(x_1, y_pred_2D, color='red', label='Regressão Linear de x_1 por y')
plt.legend(prop={'size': 12})
plt.title('Regressão Linear')
plt.xlabel('x_1: Horas diárias de estudo.',fontsize=14)
plt.ylabel('y: Nota na avaliação.',fontsize=14)
#plt.text( 2, 1.1, f'y = {lr2D.coef_[0]:.2f}x + {lr2D.intercept_:.2f}', color='red', fontsize=16)


plt.show()

```

```python

# Regressão linear para os dados 2D:
lr2D = LinearRegression()
lr2D.fit(x_2.reshape(-1, 1), y)
y_pred_2D = lr2D.predict(x_2.reshape(-1, 1))

# Plotando a regressão linear 2D:
plt.figure(figsize=(8,4))
plt.scatter(x_2, y, color='blue', label='Dados')
plt.plot(x_2, y_pred_2D, color='red', label='Regressão Linear de x_2 por y')
plt.legend(prop={'size': 12})
plt.title('Regressão Linear')
plt.xlabel('x_2: Número de advertências',fontsize=12)
plt.ylabel('y: Nota na avaliação.',fontsize=12)
plt.text( 3.8, 0, f'y = {lr2D.coef_[0]:.2f}x + {lr2D.intercept_:.2f}', color='red', fontsize=16)


plt.show()
```

```python
x = np.column_stack([x_1, x_2])
# Realizando a regressão linear
lr = LinearRegression().fit(x, y)

x_line_1 = np.array([x_1.min(), x_1.max()]) # x_1 valores
x_line_2 = np.array([x_2.min(), x_2.max()]) # x_2 valores
x_1_mean = np.full(x_line_2.shape, x_1.mean())  # aqui usamos o valor médio de x_1
x_2_mean = np.full(x_line_1.shape, x_2.mean())  # aqui usamos o valor médio de x_2
x_line_3d_1 = np.column_stack([x_line_1, x_2_mean])
x_line_3d_2 = np.column_stack([x_1_mean, x_line_2])
y_line_1 = lr.predict(x_line_3d_1)
y_line_2 = lr.predict(x_line_3d_2)

# Criando a superfície de regressão
x_surf = np.linspace(x_1.min(), x_1.max(), 100)
y_surf = np.linspace(x_2.min(), x_2.max(), 100)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)
x_surf_3d = np.column_stack([x_surf.ravel(), y_surf.ravel()])
y_surf_pred = lr.predict(x_surf_3d).reshape(x_surf.shape)

# Criando o gráfico 3D
fig = plt.figure(figsize=(5,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x_1, x_2, y, color='blue', label='Dados')
ax.plot_surface(x_surf, y_surf, y_surf_pred, color='None', alpha=0.3)
ax.plot(x_line_1, x_2_mean, y_line_1, color='red', linewidth=2, label='Regressão de x_1 por y')

ax.plot(x_1_mean, x_line_2, y_line_2, color='green', linewidth=2, label='Regressão de x_2 por y')

ax.set_title('Regressão Linear')
ax.set_xlabel('x_1: Horas diárias de estudo',fontsize=12)
ax.set_ylabel('x_2: Número de advertências.',fontsize=12)
ax.set_zlabel('y: Nota na avaliação',fontsize=12)
ax.text(5, 0, 30, f'y = {lr.coef_[0]:.2f}x_1 + {lr.coef_[1]:.2f}x_2 + {lr.intercept_:.2f}', color='red')
ax.legend()
plt.show()
```

```python
# valores x_1 para a linha vermelha, mantendo x_2 constante em seu valor médio
x_line_1 = np.array([x_1.min(), x_1.max()])  
# valores x_2 para a linha verde, mantendo x_1 constante em seu valor médio
x_line_2 = np.array([x_2.min(), x_2.max()])  

# calcular as previsões y para a linha vermelha e a linha verde
y_line_1 = lr.predict(np.array([[x_1.min(), x_2.mean()], [x_1.max(), x_2.mean()]]))
y_line_2 = lr.predict(np.array([[x_1.mean(), x_2.min()], [x_1.mean(), x_2.max()]]))

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111, projection='3d')

# plotando os pontos de dados
ax.scatter(x_1, x_2, y, color='blue', label='Data points')

# plotando a superfície de regressão
ax.plot_surface(x_surf, y_surf, y_surf_pred, color='None', alpha=0.3)

# plotando as linhas de regressão
ax.plot(x_line_1, [x_2.mean(), x_2.mean()], y_line_1, color='red', linewidth=2, label='x1-y regression at mean x2')
ax.plot([x_1.mean(), x_1.mean()], x_line_2, y_line_2, color='green', linewidth=2, label='x2-y regression at mean x1')

ax.set_title('3D Linear Regression')
ax.set_xlabel('x_1')
ax.set_ylabel('x_2')
ax.set_zlabel('y')
ax.text(5, 0, 30, f'y = {lr.coef_[0]:.2f}x_1 + {lr.coef_[1]:.2f}x_2 + {lr.intercept_:.2f}', color='red')
ax.legend()
plt.show()
```

```python


# Aumenta o tamanho da fonte
plt.rcParams.update({'font.size': 16})

# Dados de entrada
x = np.array([0.2,0.3,0.5,1.5,1.5,2.1,2.4,3.3,3.5,3.9])

y = np.array([0, 0, 0, 0, 1, 0, 1, 1, 1, 1])



# Reforça os dados para a regrssão logística ficar mais adequada aos fins didáticos

xr = reforca(x,50).reshape(-1,1)
yr = reforca(y,50)


print(len(xr))
print(len(yr))


# Regressão linear
lin_reg = LinearRegression().fit(xr, yr)
lin_pred = lin_reg.predict(xr)

# Regressão logística
log_reg = LogisticRegression().fit(xr, yr)

# Gerando um conjunto de pontos x mais detalhado para suavizar a curva
x_smooth = np.linspace(xr.min(), xr.max(), 500).reshape(-1,1)
log_proba_smooth = log_reg.predict_proba(x_smooth)[:,1]

# Plotando os resultados
plt.figure(figsize=(9, 5))
plt.scatter(xr, yr, color='blue', label='Dados originais')
plt.plot(xr, lin_pred, color='red', label='Regressão linear')
plt.plot(x_smooth, log_proba_smooth, color='green', label='Regressão logística')
plt.xlabel('Horas de estudo por dia')
plt.ylabel('Situação de aprovação')
plt.legend()
plt.show()
```

```python
b0=log_reg.intercept_[0]
b1=log_reg.coef_[0][0]
print('beta_0= ',b0)
print('beta_1= ',b1)

xp = 1.5
yp = 1/(1+np.exp(-(b0+b1*xp)))

print('p(',xp,')= ',yp)

# Plotando os resultados
plt.figure(figsize=(9, 5))
plt.scatter(xr, yr, color='blue', label='Dados originais')
#plt.plot(xr, lin_pred, color='red', label='Regressão linear')
plt.plot(x_smooth, log_proba_smooth, color='green', label='Regressão logística')
plt.xlabel('Horas de estudo por dia')
plt.ylabel('Situação de aprovação')
plt.legend()


plt.axvline(x=xp, color='grey', linestyle='--', label='linha vertical')
plt.axhline(y=yp, color='grey', linestyle='--', label='linha horizontal')
plt.scatter(xp, yp, color='red', label='ponto arbitrário', zorder=5) 


plt.show()
```
